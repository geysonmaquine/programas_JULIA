```julia
using PyPlot,LaTeXStrings
```

```julia
include("Dual_Sierpinski_Julia.jl") # rede multicamadas dual sierpinski gasket replicada em um dendrimero
include("autovalores_analiticos_DSG.jl")
include("medidas.jl")


t=collect(0:0.01:10)
fim=collect(101:1:10000)
meio=collect(11:0.1:100)
append!(meio,fim)
append!(t,meio) #tempo


autovalores1, autovetores1=eigen(LAPLACIANA_MULTICAMADA(3,1,1))
a1=multicamada_analitico(3,1,1)
base1=Matrix{Int64}(I,length(autovalores1),length(autovalores1))

autovalores2, autovetores2=eigen(LAPLACIANA_MULTICAMADA(1,3,1))
a2=multicamada_analitico(1,3,1)
base2=Matrix{Int64}(I,length(autovalores2),length(autovalores2))

autovalores3, autovetores3=eigen(LAPLACIANA_MULTICAMADA(1,1,3))
a3=multicamada_analitico(1,1,3)
base3=Matrix{Int64}(I,length(autovalores3),length(autovalores3))


autovalores4, autovetores4=eigen(LAPLACIANA_MULTICAMADA(2,2,2))
a4=multicamada_analitico(2,2,2)
base4=Matrix{Int64}(I,length(autovalores4),length(autovalores4))
```

# Probabilidade Média de Retorno Clássico

## Autovalores de L

```julia
P1=prob_med_classico(t,round.(autovalores1,digits=2))
P2=prob_med_classico(t,round.(autovalores2,digits=2))
P3=prob_med_classico(t,round.(autovalores3,digits=2))
P4=prob_med_classico(t,round.(autovalores4,digits=2))

m1=round(minimum(P1),digits=3)
m2=round(minimum(P2),digits=3)
m3=round(minimum(P3),digits=3)
m4=round(minimum(P4),digits=3)

PyPlot.loglog(t,P1,label="\$g_s=\$ 3, \$g_d=\$ 1, \$L=\$ 1 \n 1/N=$m1",color="black")
PyPlot.loglog(t,P2,label="\$g_s=\$ 1, \$g_d=\$ 3, \$L=\$ 1\n 1/N=$m2",color="green")
PyPlot.loglog(t,P3,label="\$g_s=\$ 1, \$g_d=\$ 1, \$L=\$ 3 \n 1/N=$m3",color="blue")
PyPlot.loglog(t,P4,label="\$g_s=\$ 2, \$g_d=\$ 2, \$L=\$ 2 \n 1/N=$m4",color="red")

PyPlot.title("Autovalores de L",fontsize=20)
PyPlot.xlabel("t",fontsize=20)
PyPlot.ylabel(L"$\bar{p}(t)$",fontsize=20)
PyPlot.grid(linestyle=":")
PyPlot.legend(loc="upper center", bbox_to_anchor=(1.2, .7 ), ncol=1)
PyPlot.savefig("prob_classica_laplaciana.jpg",dpi=300, bbox_inches = "tight")

```

## Autovalores Analiticos

```julia
P1=prob_med_classico(t,a1)
P2=prob_med_classico(t,a2)
P3=prob_med_classico(t,a3)
P4=prob_med_classico(t,a4)

m1=round(minimum(P1),digits=3)
m2=round(minimum(P2),digits=3)
m3=round(minimum(P3),digits=3)
m4=round(minimum(P4),digits=3)

PyPlot.loglog(t,P1,label="\$g_s=\$ 3, \$g_d=\$ 1, \$L=\$ 1 \n 1/N=$m1",color="black")
PyPlot.loglog(t,P2,label="\$g_s=\$ 1, \$g_d=\$ 3, \$L=\$ 1\n 1/N=$m2",color="green")
PyPlot.loglog(t,P3,label="\$g_s=\$ 1, \$g_d=\$ 1, \$L=\$ 3 \n 1/N=$m3",color="blue")
PyPlot.loglog(t,P4,label="\$g_s=\$ 2, \$g_d=\$ 2, \$L=\$ 2 \n 1/N=$m4",color="red")

PyPlot.title("Autovalores Analíticos",fontsize=20)

PyPlot.xlabel("t",fontsize=20)
PyPlot.ylabel(L"$\bar{p}(t)$",fontsize=20)
PyPlot.grid(linestyle=":")
PyPlot.legend(loc="upper center", bbox_to_anchor=(1.2, .7 ), ncol=1)
PyPlot.savefig("prob_classica_analitica.jpg",dpi=300, bbox_inches = "tight")
```

# Probabilidade Médio de Retorno Quântico

## Autovalores de L

```julia
PI1=prob_media_retorno_quantico(t,base1,round.(autovalores1,digits=2),autovetores1)
PI2=prob_media_retorno_quantico(t,base2,round.(autovalores2,digits=2),autovetores2)
PI3=prob_media_retorno_quantico(t,base3,round.(autovalores3,digits=2),autovetores3)
PI4=prob_media_retorno_quantico(t,base4,round.(autovalores4,digits=2),autovetores4)

PyPlot.loglog(t,PI4,label="\$g_s=\$ 2, \$g_d=\$ 2, \$L=\$ 2",color="red")
PyPlot.loglog(t,PI1,label="\$g_s=\$ 3, \$g_d=\$ 1, \$L=\$ 1 ",color="black")
PyPlot.loglog(t,PI3,label="\$g_s=\$ 1, \$g_d=\$ 1, \$L=\$ 3 ",color="blue")

PyPlot.loglog(t,PI2,label="\$g_s=\$ 1, \$g_d=\$ 3, \$L=\$ 1",color="green")


PyPlot.title("Autovalores e autovetores de L",fontsize=20)
PyPlot.xlabel("t",fontsize=20)
PyPlot.ylabel(L"$\bar{\pi}(t)$",fontsize=20)
PyPlot.grid(linestyle=":")
PyPlot.legend(loc="upper center", bbox_to_anchor=(1.2, .7 ), ncol=1)
PyPlot.savefig("prob_quantica_laplaciana.jpg",dpi=300, bbox_inches = "tight")
```

# Valor Médio e Limite Inferior

```julia
X1=valor_med_retorno(base1,round.(autovalores1,digits=2),autovetores1)
XA1=valor_med_retorno_limite_inferior(a1)


X2=valor_med_retorno(base2,round.(autovalores2,digits=2),autovetores2)
XA2=valor_med_retorno_limite_inferior(a2)

X3=valor_med_retorno(base3,round.(autovalores3,digits=2),autovetores3)
XA3=valor_med_retorno_limite_inferior(a3)


X4=valor_med_retorno(base4,round.(autovalores4,digits=2),autovetores4)
XA4=valor_med_retorno_limite_inferior(a4)

PyPlot.bar("g_s=3, \n g_d=1, \n L=1",X1,color="darkblue",label=L"$\bar{\chi}$")
PyPlot.text("g_s=3, \n g_d=1, \n L=1",(X1+0.0036),"$(string(round(X1,digits=3)))",horizontalalignment="center")
PyPlot.bar("g_s=3, \n g_d=1, \n L=1",XA1,color="gray",label=L"$\bar{\chi}^*$")
PyPlot.text("g_s=3, \n g_d=1, \n L=1",(XA1-0.01),"$(string(round(XA1,digits=3)))",horizontalalignment="center")





PyPlot.bar("g_s=1, \n g_d=3, \n L=1",X2,color="darkblue")
PyPlot.text("g_s=1, \n g_d=3, \n L=1",(X2+0.0034),"$(string(round(X2,digits=3)))",horizontalalignment="center")
PyPlot.bar("g_s=1, \n g_d=3, \n L=1",XA2,color="gray")
PyPlot.text("g_s=1, \n g_d=3, \n L=1",(XA2-0.01),"$(string(round(XA2,digits=3)))",horizontalalignment="center")


PyPlot.bar("g_s=1, \n g_d=1, \n L=3",X3,color="darkblue")
PyPlot.text("g_s=1, \n g_d=1, \n L=3",(X3+0.0035),"$(string(round(X3,digits=3)))",horizontalalignment="center")
PyPlot.bar("g_s=1, \n g_d=1, \n L=3",XA3,color="gray")
PyPlot.text("g_s=1, \n g_d=1, \n L=3",(XA3-0.01),"$(string(round(XA3,digits=3)))",horizontalalignment="center")


PyPlot.bar("g_s=2, \n g_d=2, \n L=2",X4,color="darkblue")
PyPlot.text("g_s=2, \n g_d=2, \n L=2",(X4+0.0035),"$(string(round(X4,digits=3)))",horizontalalignment="center")
PyPlot.bar("g_s=2, \n g_d=2, \n L=2",XA4,color="gray")
PyPlot.text("g_s=2, \n g_d=2, \n L=2",(XA4-0.01),"$(string(round(XA4,digits=3)))",horizontalalignment="center")



PyPlot.grid(linestyle=":")
PyPlot.xlabel("Estruturas",fontsize=20)
PyPlot.ylabel(L"$\bar{\chi},\bar{\chi}^*$",fontsize=20)
PyPlot.legend(loc="upper right")

PyPlot.savefig("valor_medio.jpg",dpi=300, bbox_inches = "tight")
```
